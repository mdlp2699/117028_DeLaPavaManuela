```python
def fib(m):
    '''
    esto realiza la serie de Fibonacci
    '''
    a=1
    b=0
    while m>b:
        a,b=b,a+b
    return m==b

#%%
def Digitos_Numero(Numero) :
    
    '''separa los digitos de un numero y los agregra a una lista vacia
    '''
    Lista=[]
    while True:
        dig=Numero%10
        Lista.append(dig)
        Numero//=10
        if Numero==0: 
            break
    return Lista
#%%

def Leer_numero():
    '''
    lee un numero entero
    '''
    while True:
        try:
            
            Num_1=int(input('digite un numero: '))
            break
        except:
           print('debe ser un numero entero')
    return Num_1
#%%   
y=Leer_numero()      

#si el numero es negativo se multiplica por -1
if y<0: y*=(-1)

x=sum(Digitos_Numero(y)) #esto suma los digitos del numero pedido

if fib(x) is True:
    print('la suma de los digitos de',y,'es un numero de Fibonacci.')
else:
    print('la suma de los digitos de',y,'no es un numero de Fibonacci.')    
```

```python
VocalesM=['A','E','I','O','U','Á','É','Í','Ó','Ú','Ü']
Letras_tilde=['á','é','í','ó','ú','Á','É','Í','Ó','Ú']
Digitos=['1','2','3','4','5','6','7','8','9','0']
Espacio=[' ']
Cont_kw=0

#%%

import keyword

#%%

def Contador_cadena(lista,cadena):
    ''' cuenta elementos comunes enre una lista y una cadenaio
    '''
    contador=0
    for i in cadena:
        if i in lista:
            contador+=1
    return contador 
#%%

Cadena=input('escriba una cadena: ')

Cont_vocM=Contador_cadena(VocalesM,Cadena)
print('la cantidad de vocales en mayuscula es: ',Cont_vocM)

Cont_tildes=Contador_cadena(Letras_tilde,Cadena)
print('la cantidad de vocales con tildes es: ', Cont_tildes)

Cont_digitos=Contador_cadena(Digitos,Cadena)
print('la cantidad de digitos es: ',Cont_digitos)

Cont_espacios=Contador_cadena(Espacio,Cadena)
print('la cantidad de espacios es: ',Cont_espacios)

y=Cadena.split() #separa palabras en una cadena

#revisa si alguna de las palabras esta dentro del modulo Keywords
for i in y:    
    if i in keyword.kwlist:
        Cont_kw+=1
print('la cantidad de palabras clave es: ',Cont_kw)
```

```python
def Leer_lista():
    '''
    lee una Lista
    '''
    while True:
        try:
            Lista=eval(input('escriba una lista: '))
            if type(Lista) is list:
                break
            else:
                print('Debe ser una lista, recuerde que tiene una estructura\
 [,]')
        except:
            print('Debe ser una lista, intentelo de nuevo.')
    return Lista
#%%
# comprueba que cada elemeto sea una lista
def Entero(Lista):
    '''
    comprueba que cada elemento sea un numero entero dentro de la lista
    '''
    for i in Lista:
        if type(i) is int:
            C=True
        else:
            C=False
            break
    return C
#%%

def Leer_entero():
    '''
    lee un numero Entero
    '''
    while True:
        try:
            Numero=int(input('Digite un numero entero: '))
            break
        except:
            print('Debe ser un numero, intentelo de nuevo.')
    return Numero
#%%
def Lista_asc(Lista):
    '''
    asegura que en la lista, sus elementos estan ordenados de forma 
    ascendente
    '''
    for i in range(0,len(Lista)-1):
        if Lista[i] <= Lista[i+1]:
            C=True
        else:
            C=False
            break
    return C
#%%

while True:
    Lista_1=Leer_lista()
    if Entero(Lista_1) == True and Lista_asc(Lista_1) == True :
        break
    print('Todos los elementos deben ser enteros y deben estar ordenados\
 ascendentemente. ')

Numero_1=Leer_entero()

for i in range(len(Lista_1)):
    if Numero_1 <= Lista_1[len(Lista_1) - 1] :  # si el numero es menor o igual
                                                # a la ultima posicion de 
                                                # la lista
                                                
        if Lista_1[i]>=Numero_1:                # si el elemento es menor o 
                                                # igual al numero
            Lista_1.insert(i,Numero_1)          # lo agrega 
            break
    else :
        Lista_1.append(Numero_1)

        
print('La nueva lista es: ',Lista_1)
```
